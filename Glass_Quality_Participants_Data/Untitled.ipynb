```python
#creating histogram graph for each variable/features
def features(i):
    if data[i].dtype!='object':
        plt.figure(figsize=(8,5))
        plt.title("histogram of "+str(i))        
        data[i].plot.hist(bins = 16)
        plt.show()
        plt.clf()
sns.set() 
j=['max_luminosity','thickness','pixel_area']
for k in j:
    features(i=k)
```

```python
plt.figure(figsize=(14,6))
sns.heatmap(data.corr(),annot=True)
plt.show
```

```python
data_copy = data.copy()
data_copy.head()
```

```python
for k in list(data_copy.columns):
    IQR=np.percentile(data_copy[k],75) - np.percentile(data_copy[k],25)
    
    Outlier_top=np.percentile(data_copy[k],75)+1.5*IQR
    Outlier_bottom=np.percentile(data_copy[k],25)-1.5*IQR
    
    data_copy[k]=np.where(data_copy[k] > Outlier_top,np.percentile(data_copy[k],50),data_copy[k])
    data_copy[k]=np.where(data_copy[k] < Outlier_bottom,np.percentile(data_copy[k],50),data_copy[k])
```

```python
plt.figure(figsize=(14,6))
sns.heatmap(data_copy.corr(),annot=True)
plt.show
```

```python
def df(i):
    if data_copy[i].dtype!='object':
        plt.show()
        plt.title("histogram of "+str(i))
        data_copy[i].plot.kde()
        plt.show()
for k in data_copy.columns:
    df(i=k)
```

```python
#data_copy = data_copy.dropna('x_component_1')
data_copy.head()
```

```python
x = data_copy.drop('class', axis=1)
y = data_copy['class']

from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.25,random_state=7)
```

```python
from sklearn.preprocessing import MinMaxScaler
sc = MinMaxScaler()

x_train = sc.fit_transform(x_train)
x_test = sc.fit_transform(x_test)
```

```python
from sklearn.linear_model import LogisticRegression
model1 = LogisticRegression()
model1.fit(x_train,y_train)
y_pred = model1.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_pred,y_test)
print(cm)
```

```python
print(model1.score(x_train,y_train))
print(model1.score(x_test,y_test))
```

```python
from sklearn.svm import SVC
model2 = SVC()
model2.fit(x_train,y_train)
y_pred = model2.predict(x_test)
print(model2.score(x_train,y_train))
print(model2.score(x_test,y_test))
```

```python
from sklearn.ensemble import RandomForestClassifier
model3 = RandomForestClassifier()
model3.fit(x_train,y_train)
y_pred = model3.predict(x_test)
print(model3.score(x_train,y_train))
print(model3.score(x_test,y_test))
```

```python
from sklearn.ensemble import GradientBoostingClassifier
model4 = GradientBoostingClassifier()
model4.fit(x_train,y_train)
y_pred = model4.predict(x_test)
print(model4.score(x_train,y_train))
print(model4.score(x_test,y_test))
```

```python
from sklearn.model_selection import KFold
kf = KFold(n_splits=7) 
```

```python
for i,j in kf.split(x):
    print(f"{i.shape} --- {j.shape}")
```

```python
train_accuracy = []
test_accuracy = []

for train_index, test_index in kf.split(x):
    
    X_train = x.loc[train_index]
    X_test = x.loc[test_index]
    y_train = y.loc[train_index]
    y_test = y.loc[test_index]
    
    train_accuracy.append(model2.score(X_train, y_train))
    test_accuracy.append(model2.score(X_test, y_test))
    
print(np.mean(np.array(train_accuracy)))
print(np.mean(np.array(test_accuracy)))
```
